```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

fig, ax = plt.subplots()
m = Basemap(projection='merc',llcrnrlat=-30,urcrnrlat=30,
            llcrnrlon=40,urcrnrlon=160,ax=ax,resolution='c')
m.fillcontinents(color='tan', zorder=0)
m.drawcoastlines(color='k', linewidth=0.7, zorder=10)
m.drawcountries(color='k', linewidth=0.5, zorder=10)
soundings.plot_sites_on_map(m, include_labels=True, real_names=False, color='green')
plt.show()
```
